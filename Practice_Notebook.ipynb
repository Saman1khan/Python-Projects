```python
plugged_in = input("Is the lamp plugged in?")
burned_out = input("Is the bulb burned out?")
```

```python
if plugged_in == 'Yes':
    if burned_out == 'Yes':
        print('Replace your Bulb')
    else:
        print('Repair the Lamp')
else:
    print("Plese plug in your Lamp")
```

```python
if plugged_in == 'No':
    print("Please plug in your Lamp")
elif plugged_in == 'Yes':
    if burned_out == 'No':
        print('Please repair your lamp')
    else:
        print('Replace your Bulb')
else:
    print('In the Else')
```

```python
a = int(input("Enter the value of A: "))

while a < 5:
    print("Current value of A:", a)
    a += 1

print("Final value of A:", a)
```

```python
correct_password = "password"
attempts = 3

input_password = input("Please enter your password: ")

while (input_password != correct_password) and (attempts > 0):
    attempts -= 1
    print("Incorrect Password. Attempts Left", attempts)
    input_password = input("Please enter the correct password: ")

if attempts > 0:
    print("Successfully Logged In!")
else:
    print("Account has been locked")
```
